```python
%%writefile app.py
import streamlit as st
import pandas as pd
from sklearn.linear_model import LinearRegression

# dataset load
data = pd.read_csv("train (1).csv")

X = data[['OverallQual','GrLivArea','GarageCars','TotalBsmtSF']]
y = data['SalePrice']

# model train
model = LinearRegression()
model.fit(X,y)

st.title("House Price Prediction")
st.write("Enter house details:")

overall = st.text_input("Overall Quality")
area = st.text_input("Living Area")
garage = st.text_input("Garage Cars")
basement = st.text_input("Basement Area")

if st.button("Predict Price"):
    try:
        house = [[float(overall), float(area), float(garage), float(basement)]]
        price = model.predict(house)
        st.success(f"Predicted Price: ${price[0]:,.2f}")
    except:
        st.error("Please enter valid numbers")
```
